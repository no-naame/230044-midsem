## Assumption 1: Discrete Node Labels

**Statement**: The WL relabeling scheme requires that nodes have **discrete labels** drawn from a finite alphabet $\Sigma$. The entire mechanism of multiset collection, sorting, string concatenation, and hash-based compression operates on discrete symbols.

**Where in the paper**: Section 1, Definition in Section 1.1 — the graph is defined as $G = (V, E, l)$ where $l: V \to \Sigma$ maps nodes to a finite label set.

**Why it matters**: The label compression step (Algorithm 1, Step 3) requires an injective mapping from strings of labels to new labels. This is straightforward for discrete labels (concatenate and hash) but does not naturally extend to continuous or high-dimensional node attributes.

**Violation scenario**: Molecular graphs where nodes (atoms) carry continuous features such as 3D spatial coordinates, partial charges, or real-valued physicochemical descriptors. These cannot be directly hashed or sorted in the same way as discrete labels, requiring discretization (which loses information) or a fundamentally different kernel formulation.

## Assumption 2: Injective Label Compression (Sufficient Label Alphabet)

**Statement**: The compression function $f$ that maps multiset-label strings to new compressed labels must be **injective** — distinct strings must map to distinct labels. The paper requires that the label alphabet size satisfies $|\Sigma| \geq 2n$ (where $n$ is the total number of nodes across the two graphs being compared) to guarantee injectivity.

**Where in the paper**: Section 2, paragraph after Algorithm 1 — "we need to ensure that the mapping $f$ is injective... the label alphabet must be large enough."

**Why it matters**: If two different multiset-label strings are mapped to the same compressed label (a hash collision), the kernel loses the ability to distinguish those neighborhood patterns. The resulting feature vectors would conflate structurally different neighborhoods, degrading classification performance.

**Violation scenario**: Extremely large graphs (millions of nodes) with many diverse neighborhood patterns could in principle exhaust the hash space and cause collisions, though in practice with dictionary-based hashing this is unlikely. A more realistic concern arises when using fixed-size hash functions (e.g., 32-bit) on graphs with a very large number of distinct neighborhood patterns.

## Assumption 3: Structural Distinguishability via the 1-Dimensional WL Test

**Statement**: The WL subtree kernel implicitly assumes that structurally different graphs will produce **different label multisets** after $h$ iterations of the 1-dimensional WL relabeling. Graphs that the 1-dim WL test cannot distinguish will receive identical feature vectors and thus a kernel value equal to their self-similarity — making classification between them impossible.

**Where in the paper**: Section 2, paragraph after Algorithm 1 — the connection between WL relabeling and the classical WL graph isomorphism test is established. The paper acknowledges that "the WL test of isomorphism... is a necessary but not sufficient condition for graph isomorphism" and references Cai, Fürer, and Immerman (1992) for examples of non-isomorphic graphs that the 1-dim WL test cannot distinguish.

**Why it matters**: This is a fundamental expressivity limitation. The WL subtree kernel can only distinguish graphs that the 1-dim WL test can distinguish. Any pair of non-isomorphic graphs that are WL-equivalent (produce the same multiset of labels at every iteration) will be indistinguishable to the kernel.

**Violation scenario**: Pairs of non-isomorphic **regular graphs** with identical degree sequences. For example, two different 3-regular graphs on the same number of nodes where all nodes have the same initial label. After each WL iteration, every node collects the same multiset of neighbor labels (since all neighbors have the same label and the same degree), so all nodes receive the same compressed label. The WL kernel produces identical feature vectors despite the graphs being structurally different. This is the classic failure case of the 1-dim WL test (Cai et al., 1992).
