## Step-by-Step Description of the Core Method

The paper proposes the **Weisfeiler-Lehman (WL) subtree kernel**, an efficient graph kernel for graph classification. The method works as follows:

### Step 1: Input Representation
Each graph is represented as $G = (V, E, l)$ where $V$ is the set of nodes, $E$ is the set of edges, and $l: V \to \Sigma$ is a function assigning discrete labels from a finite alphabet $\Sigma$ to each node. (Section 1, Definition in Section 1.1)

### Step 2: Multiset-Label Determination
For each node $v$, collect the **multiset** of its neighbors' labels:
$$M_i(v) = \{l_{i-1}(u) \mid u \in N(v)\}$$
where $N(v)$ is the set of neighbors of $v$ and $l_{i-1}$ are the labels from the previous iteration. (Algorithm 1, Step 1)

### Step 3: Sorting and String Creation
Sort the elements of $M_i(v)$ to obtain a canonical ordering, then concatenate them into a string $s_i(v)$. **Prepend** the node's own previous label $l_{i-1}(v)$ as a prefix to this string. This ensures that two nodes with different labels but identical neighborhoods receive different compressed labels. (Algorithm 1, Step 2)

### Step 4: Label Compression
Map each unique string $s_i(v)$ to a new compressed integer label using an **injective function** $f$. In practice, this is implemented as a hash function (dictionary lookup) that assigns a new integer to each previously unseen string. (Algorithm 1, Step 3; hash function in Algorithm 2)

### Step 5: Relabeling
Set $l_i(v) := f(s_i(v))$ for all nodes $v$ in all graphs. This completes one WL iteration. (Algorithm 1, Step 4)

### Step 6: Feature Vector Construction
After $h$ iterations, construct the feature vector $\phi_{\text{WLsubtree}}(G)$ by counting occurrences of **all labels** across all iterations (original labels at $i=0$ plus compressed labels at $i=1, \ldots, h$). This produces a histogram over the entire label vocabulary. (Definition 4, Equation 2, Figure 2)

### Step 7: Kernel Computation
The WL subtree kernel between two graphs is the **dot product** of their feature vectors:
$$K_{\text{WLsubtree}}(G, G') = \langle \phi_{\text{WLsubtree}}(G), \phi_{\text{WLsubtree}}(G') \rangle$$
(Equation 2)

### Step 8: Classification
Feed the kernel matrix $K$ into an SVM with a precomputed kernel for graph classification. The SVM hyperparameter $C$ is optimized via cross-validation. (Section 4.2.2)

## Summary

The paper solves the problem of **efficient graph classification** by proposing the WL subtree kernel. It captures subtree patterns via iterative neighborhood label aggregation: at each iteration, a node's label encodes the structure of its depth-$i$ neighborhood. The resulting feature vectors (histograms of all labels across iterations) are compared via dot product to define a positive semi-definite kernel.

The key advantage is **computational efficiency**: the WL subtree kernel runs in $O(hm)$ time, where $h$ is the number of iterations and $m$ is the number of edges. This is **linear in the number of edges**, making it scalable to large labeled graphs where prior graph kernels had $O(n^3)$ or worse complexity. On benchmark datasets, it achieves state-of-the-art or competitive classification accuracy while being orders of magnitude faster than existing methods.
